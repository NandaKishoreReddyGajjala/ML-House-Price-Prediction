#### we can observe that the data is right skewed which could not an ideal method of using the data, so we need to make the data into gausian curve.

```python
train_data.corr(numeric_only=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
sns.heatmap(train_data.select_dtypes(include=['number']).corr(), annot=True, cmap="YlGnBu")
```

#### Transforming the data using LOG to get the gausian curve

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'])
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'])
train_data['population'] = np.log(train_data['population'])
train_data['households'] = np.log(train_data['households'])
```

```python
train_data.hist(figsize=(15,10))
```

```python
train_data.corr(numeric_only=True)
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop('ocean_proximity', axis=1)
train_data
```

```python
train_data['INLAND'] = train_data['INLAND'].astype(int)
train_data
```

```python
columns_to_convert = ['<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY', 'NEAR OCEAN']
train_data[columns_to_convert] = train_data[columns_to_convert].astype(int)
```

```python
train_data
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(train_data.select_dtypes(include=['number']).corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```
